```python
import pandas as pd
import matplotlib.pyplot as plt
import pathlib
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from urllib import request
```

```python
datasets = pathlib.Path().cwd().parents[2] / 'data/sample-data'
datasets.mkdir(parents=True, exist_ok=True)
```

```python
data_root = 'https://github.com/ageron/data/raw/main/' + 'lifesat/'
```

```python
sample_files = ['lifesat.csv', "oecd_bli.csv", "gdp_per_capita.csv"]
```

```python
for file in sample_files:
    if not (datasets / file).is_file():
        print(f'Downloding the file: {file}')
        url = data_root + file
        print(url)
        datapath = datasets / file
        request.urlretrieve(url, datapath)
        
```

```python
file_name = 'lifesat.csv'
```

```python
lifestat = pd.read_csv(datasets / file_name)
```

```python
lifestat.plot(kind='scatter', x = 'GDP per capita (USD)', y = 'Life satisfaction', grid = True  )
plt.axis([25000, 62000, 5, 8])
plt.show()
```

```python
model = LinearRegression()
```

```python
X = lifestat[['GDP per capita (USD)']].values
y = lifestat[['Life satisfaction']].values
```

```python
model.fit(X,y)
```

```python
X_new = [[38123]]
```

```python
y_pred = model.predict(X_new)
```
